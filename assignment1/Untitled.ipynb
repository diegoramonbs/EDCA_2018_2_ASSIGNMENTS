```python
%matplotlib inline
import matplotlib.pyplot as plt 
import timeit 
from functools import partial
import random
from algorithm import Sort
```

```python
# Based on: https://electronut.in/plotting-algorithmic-time-complexity-of-a-function-using-python/
def plot_time(fn, nMin, nMax, nInc, nTests):
    x = []
    y = []
    for i in range(nMin, nMax, nInc):
        N = [random.randint(1,1000) for _ in range(i)]
        testNTimer = timeit.Timer(partial(fn, N))
        t = testNTimer.timeit(number=nTests)
        x.append(i)
        y.append(t)
    p1 = plt.plot(x, y, '-', label=fn.__name__)
```

```python
plot_time(Sort.insertion_sort, 10, 1000, 1, 10)
plot_time(Sort.selection_sort, 10, 1000, 1, 10)
plot_time(Sort.merge_sort, 10, 1000, 1, 10)
plt.legend()
```
